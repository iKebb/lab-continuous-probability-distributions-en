## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

$$ f(x) = \frac{1}{b - a} \quad \text{for} \quad a \leq x \leq b $$


```python
def uniform_pdf(x, a, b):
  if a <= x <= b:
    return 1 / (b-a)
  else:
    return 0
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate

1. Call your function with 2 sets of params below:

```python
def generate_uniform_numbers(a, b, x):
  return stats.uniform.rvs(size=x, loc=a, scale=b-a)
```

```python
a= 10
b= 15
x= 100

case_1= generate_uniform_numbers(a, b, x)
```

```python
a= 10
b= 60
x= 1000

case_2= generate_uniform_numbers(a, b, x)
```

```python
plt.hist(case_1, bins=30, density=True, alpha=0.7, color='orange')
plt.title('Uniform Distribution [$a=10$, $b=15$]')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.hist(case_2, bins=30, density=True, alpha=0.7, color='orange')
plt.title('Uniform Distribution [$a=10$, $b=60$]')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
def norm_distribution(mu, sigma, n):
  return stats.norm.rvs(size=n, loc=mu, scale=sigma)
```

```python
n= 1000
mu, sigma= 10, 1

case_1= norm_distribution(mu, sigma, n)
```

```python
n= 1000
mu, sigma= 10, 50

case_2= norm_distribution(mu, sigma, n)
```

```python
# Plotting the sample distribution
plt.hist(case_1, bins=30, density=True, alpha=0.6, color='slateblue')
plt.title('Normal Distribution [$\mu=0$, $\sigma=1$]')
plt.xlabel('Value')
plt.ylabel('Density')

# Plotting the PDF of the Normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

plt.show()
```

```python
# Plotting the sample distribution
plt.hist(case_2, bins=30, density=True, alpha=0.6, color='slateblue')
plt.title('Normal Distribution [$\mu=0$, $\sigma=1$]')
plt.xlabel('Value')
plt.ylabel('Density')

# Plotting the PDF of the Normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

plt.show()
```

How are the two distributions different?

Naturally, the standard deviation is way more dense, we can see it
